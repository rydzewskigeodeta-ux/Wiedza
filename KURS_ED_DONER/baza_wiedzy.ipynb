### Kroki

1. pobieranie repo `git clone ...`
2. w GitHubie robienie `Fork`
3. u siebie w powershell'u `git remote -v` następnie pojawi się:

    > origin  https://github.com/ed-donner/agents.git (fetch)
    > 
    > origin  https://github.com/ed-donner/agents.git (push)

4. dodajemy nowy pilot `git remote rename origin upstream` -> zamieniamy tym projekt aby nie zmieniać tego repo tylko nasze na naszym forku tzn. nasz będzie origin a Eda upstream
Kolejne kroki to:

```
        git remote add origin https://github.com/<your‑username>/agents.git 
        git remote -v
```
> gdzie `https://github.com/<your‑username>/agents.git` z mojego forka
##### Wynik to oznacza, że mój jest origin, a Eda upstream
   > origin    https://github.com/<your‑username>/agents.git (fetch) 
   > 
   > origin    https://github.com/<your‑username>/agents.git (push)
   > 
    > upstream  https://github.com/ed-donner/agents.git (fetch)
    > 
    > upstream  https://github.com/ed-donner/agents.git (push)


### Co to znaczy? `upstream` to jest do pobierania zmian z kursu, a `origin` ja nanim działam i jest orginałem moim.
