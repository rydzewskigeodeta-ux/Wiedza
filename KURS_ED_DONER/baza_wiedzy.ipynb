## Kroki

### Rozdzielanie repozytorium
1. pobieranie repo `git clone ...`
2. w GitHubie robienie `Fork`
3. u siebie w powershell'u `git remote -v` następnie pojawi się:

    > origin  https://github.com/ed-donner/agents.git (fetch)
    > 
    > origin  https://github.com/ed-donner/agents.git (push)

4. dodajemy nowy pilot `git remote rename origin upstream` -> zamieniamy tym projekt aby nie zmieniać tego repo tylko nasze na naszym forku tzn. nasz będzie origin a Eda upstream
Kolejne kroki to:

```
        git remote add origin https://github.com/<your‑username>/agents.git 
        git remote -v
```
> gdzie `https://github.com/<your‑username>/agents.git` z mojego forka
##### Wynik to oznacza, że mój jest origin, a Eda upstream
   > origin    https://github.com/<your‑username>/agents.git (fetch) 
   > 
   > origin    https://github.com/<your‑username>/agents.git (push)
   > 
    > upstream  https://github.com/ed-donner/agents.git (fetch)
    > 
    > upstream  https://github.com/ed-donner/agents.git (push)


##### Co to znaczy? `upstream` to jest do pobierania zmian z kursu, a `origin` ja nanim działam i jest orginałem moim. 

#### Przygotowanie wkładu

Pliki dodajemy w VS Code lub Cursor przez przeciągnięcie lub założenie plików. W projekcie są już foldery o nazwie `community_contributions` i do nich należy przesłać pliki lub założyć.

<h4 style="color:red"> Zawsze pracuj z gałęzią, która opisuje zmianę:</h4>
 
 > Przykład opisu gałęzi `LandSur_PL/day1`
 >
 > Komenda/polecenie:
 >
 > `git checkout -b LandSur_PL/day1`

#### Dodawanie i zatwierdzanie

1. sprawdzaj status komendą:<br>
    `git status`
    > Wynik:  
    > Powinieneś zobaczyć swoje nowe/zmodyfikowane pliki w Untracked files lub Modified.

2. Sprawdź czy plik został przeniesiony do zatwierdzenia zmiany komendami 
    dla pojedyńczych plików <span style="color:red"> ZALECANE </span>:<br>
    `git add section1/community_contributions/nazwa_mojego_pliku.ipynb` <br> `git status`

    dla wielu plików: <br>
    `git add .\community_contributions\`

3. Wykonaj commit i sprawdź 

    `git commit -m "Add LandSur_PL's lab 1 solution (cleared outputs)"`<br>
    `git status`
    > Wynik:  
    > On branch LandSur_PL/day1 nothing to commit, working tree clean  
4. Wypchnij dane do widelca `fork` i otwórz Pull Request poleceniem:  
    `git push -u origin LandSur_PL/day1`

> 4.Kliknij Create pull request.  

#### Utrzymuj widelec `fork` w synchronizacji w pryszłych pracach  

1. Aktualizacja wykonuj następujące komendy:  
    `git checkout main`  -> włącz main  
    `git pull upstream main`  ->  pobierz z lokalnego do głównego  
    `git push origin main `   ->  wypchnij wyślij do mojego forka  
2. Mój kolejny wkład:  
    `git checkout -b LandSur_PL/nastepna_zmiana main`
    pracuj ...
    commituj
    wypychaj 
    otwieraj PR 
    ...

## Błędy i problemy  

| Objawy                                                      | Przyczyna                                       | Rozwiązanie                                                                                                                                                      |
| ------------------------------------------------------------- | -------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Pliki spoza folderu społeczności pojawiają się w PR**               | Inscenizowane przypadkowo                           | `bash\n# unstage\ngit restore --staged path/to/file\n# optionally discard changes\ngit checkout -- path/to/file\n` Naciśnij ponownie; PR aktualizuje się automatycznie. |
| **Notes pokazuje duże czerwono-zielone bloki JSON**             | Wyniki nie zostały zatwierdzone                     | Czyść wyjścia → Zapisz → `git add` zeszytu ponownie →  `git commit --amend` (lub nowym commit) → `git push`.                                           |
| **Git odmawia napięcia (bez przewijania do przodu) (`non‑fast‑forward`)**                  | Lokalna gałąź za zdalną	                   | `git pull --rebase origin alice/lab1-solution` To pchaj                                                                                                |
| **`fatal: remote origin already exists` przy dodawaniu zdalnych kontrolerów PR** | Już załatwiłeś                       | Użyj `git remote -v` do potwierdzenia adresów URL;                                                                                                  |
| **Przypadkowo usunąłem plik w PR**                     | Deklaracja jest częścią zobowiązania	               | Przywróć plik lokalnie, dodaj, zatwierdz, wypchnij – GitHub zaktualizuje PR, aby pokazać, że usunięcie zniknęło.                                                          |
| **Trzeba usunąć cały plik z repozytorium i PR**              | Plik nie powinien być śledzony (np. duże dane) | `bash\ngit rm --cached path/to/big.csv\n.gitignore it if necessary\n` Zatwierdzaj się i pchaj.                                                                     |
