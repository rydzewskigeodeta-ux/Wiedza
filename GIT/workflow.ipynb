# Workflow dla GitHuba



#### SCHEMAT
---
|Przepływ          |Komendy     / co się dzieje        |
|------------------|--------------------|
|Start (stabilny kod)|`git status` <br> `git branch` <br> `git checkout main` <br>`git pull` |
|↓|sprawdzasz status <br> sprawdzasz gałezie <br> uruchamiasz gałąź `main` główną może być `master` <br> pobierasz dane z repo |
|Tworzysz branch nową gałąź `nazwa_galezi`|`git checkout -b [nazwa_galezi]` <br> ----- <br> można również: <br> `git branch [nazwa_galezi]` <br> `git checkout [nazwa_galezi]`|
|↓|tworzysz i aktywujesz gałąż `[nazwa_galezi]`|
|Pracujesz + commitujesz|Piszesz ... robisz checkpoint -> <br> `git add .` <br> `git commit -m "opis_zmainy"` <br> `git push -u origin [nazwa_galezi]` (pierwszy raz) potem <br> `git push` |
|↓|Praca ... chekpoint <br> zapisujesz wszystke plki i foldery <br> opisujesz zmianę <br> wysyłasz do GiHuba|
|Kończysz pracę -> wracasz do main|`git checkout main` -> `main` - główna gałąź sprawdż 'git branch'<br> `git pull` <br> `git pull --rebase`|
|↓|wracasz do głównej gałęzi <br> pobierasz nowe zmiany jak robisz sam <br> bezpieczniejsza wersja i dla osób pracujących z innymi|
|MERGE (łączysz zmiany) |`git merge [nazwa_galezi]` <br> `git push`|
|↓|połączyłeś zmiany|
|Kończysz gałąź ->|`git branch -d [nazwa_galezi]` <br> `git push origin --delete [nazwa_galezi]`| 
|↓|usuwasz gałąź w projekcie <br> usuwasz gałąź z repo na GitHubie|
|Koniec AKTUALNA BAZA !!!||

#### OPIS
---

1.Budujesz gita

2.Praca:

2.1. Rozpoczęcie pracy nad częścią projektu (`XYZ` - nazwa gałęzi)

   a) tworzysz gałąź -> `git checkout -b XYZ` (`XYZ` - nazwa gałęzi)

   b) dodajesz zmiany -> `git add .`

   c) zapisujesz -> `git commit -m "opis_zmiany"` wykonujesz je często opisując dokładnie zmiany

   d) wysyłasz gałąź -> `git push -u origin XYZ` (`XYZ` - XYZ gałęzi) 

2.2. Koniec pracy nad gałęzią

   a) wracasz do głównej gałęzi -> `git checkout main` (`main` - główna gałąź)

   b) aktualizujesz zmiany `git pull`

2.3. Łączysz co zrobiłeś w gałęzi `XYZ`

   `git merge XYZ` 

2.4. Wysyłasz do  gita
 
   `git push`

2.5. Usuwanie gałęzi

   a) usuwasz gałąź lokalnie `git branch -d XYZ` (`XYZ`- nazwa gałezi)

   b) usuwasz gałąźz GitHuba `git push origin --delete XYZ` (`XYZ`- nazwa gałezi)

# Workflow dla GitHuba

#### __Założenie projektu__
| Krok | Komenda                     | Co robi                   |
| ---- | --------------------------- | ------------------------- |
| 1    | `git init`                  | Inicjalizuje repozytorium |
| 2    | `git add .`                 | Dodaje wszystkie pliki    |
| 3    | `git commit -m "init"`      | Pierwszy zapis projektu   |
| 4    | `git branch -M main`        | Ustawia główną gałąź      |
| 5    | `git remote add origin URL` | Podłącza GitHub `URL = link do repo z GitHub`    |
| 6    | `git push -u origin main` lub `git push`  | Wysyła projekt do GitHub  |

#### __Codzienny przepływ pracy__
🔹 Start pracy

| Komenda             | Co robi                  |
| ------------------- | ------------------------ |
| `git status`        | sprawdza stan projektu   |
| `git pull`          | pobiera najnowsze zmiany |
| `git branch`        | pokazuje gałęzie         |
| `git checkout main` | przełącza na main        |

🔹 Tworzenie nowej funkcji

| Komenda                         | Co robi                     |
| ------------------------------- | --------------------------- |
| `git checkout -b XYZ` | tworzy i przełącza na gałąź |
| `git status`                    | kontrola zmian              |


🔹 W trakcie pracy
| Komenda                | Co robi          |
| ---------------------- | ---------------- |
| `git add .`            | dodaje zmiany    |
| `git commit -m "opis"` | zapisuje zmiany  |
| `git push`             | wysyła do GitHub |

TIP: 💡
 > rób małe commity

 > opisuj konkretnie


| Komenda                       | Co robi          |
| ----------------------------- | ---------------- |
| `git checkout main`           | wraca na main    |
| `git pull`                    | aktualizuje main |
| `git merge XYZ`               | łączy zmiany     |
| `git push`                    | wysyła zmiany    |
| `git branch -d XYZ`           | usuwa gałąź      |
