# <span style="color:blue">KOMENDY DO GIT-a </span>

|Komendy|Co robi|TIP|
|---|---|---|
|```git init```|inicjalizuje repo lokalne na komputerze||
|```git remote add origin <URL>```|podpina repo z GitHuba|<URL> - skopiuj z githuba po założeniu repo|
|```git commit -m "init"```|zapisuje pierwszy stan projektu nazwa="init"|"init" oznacza nazwę zmiany opisuje co zrobiłeś|
||||
