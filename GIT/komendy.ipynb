## <span style="color:green">___KOMENDY DO GIT-a___ </span>
(#UV_wiedza)
|Komendy|Co robi|TIP/szczegóły|
|-----------------------------------|-----------------------------------------------|---|
|```git init```                     |inicjalizuje repo lokalne na komputerze        ||
|```git remote add origin <URL>```  |podpina repo z GitHuba                         |```<URL>``` - skopiuj z githuba po założeniu repo|
|```git commit -m "init"```         |zapisuje pierwszy stan projektu nazwa="init"   |"init" oznacza nazwę zmiany opisuje co zrobiłeś|
|```git add .```                    |dodaje wszystkie zmniany do zapisu             |```.``` oznacza że wszystkie jeżeli dodasz ```nazwa_pliku.txt``` dotyczy tylko pliku|
|```git push```                     |wysyła projekt na _Githuba_                    ||
|```git branch```                   |pokazuje jaka jest gałąź                       |domyśnie gałęzią jest <span style="color:green">*master*</span>|
|```git diff```                     |pokazuje wszystko co zmieniłeś w plikach np. kod||
|```git log```                      |pokazuje historie zmian||
|```git fletch```                   |sprawdza zmiany, ale ich nie wprowadza||
|```git status```                   |pokazuje pliki które uległy zmianie||

---
🔹 Podstawowe
| Komenda         | Opis          | Tip                  |
| --------------- | ------------- | -------------------- |
| `git init`      | start repo    | tylko raz            |
| `git clone URL` | pobiera repo  | zamiast init         |
| `git status`    | stan repo     | używaj często        |
| `git add .`     | dodaje zmiany | ostrożnie (wszystko) |
| `git commit -m` | zapis zmian   | opisuj dobrze        |

🔹 Gałęzie

| Komenda                | Opis            | Tip                 |
| ---------------------- | ----------------------------------- | ------------------- |
| `git branch`           | lista gałęzi                         |                     |
| `git branch name`      | tworzy gałąź `name` nazwa gałęzi    |                     |
| `git checkout name`    | zmiana gałęzi `name` nazwa gałęzi   |                     |
| `git checkout -b name` | tworzy i przełącza do tej gałęzi `name` nazwa gałęzi | najczęściej używane |
| `git branch -d name`   | usuwa gałąź `name` nazwa gałęzi     | po merge            |
