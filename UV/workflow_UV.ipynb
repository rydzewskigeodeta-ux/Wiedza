| ***Komenda***|***Co robi***|
|-----------------|------------------------------------------|
|*uv --version*|jaka wersja UV|
|*uv self update*|sprawdza aktualizacje|
| *uv init [nazwa_projektu]*|założenie projektu|
|*uv venv*|Tworzy środowisko w .venv/ (izolowany Python dla projektu)|
|*uv add ipykernel*|Instaluje kernel Jupyter w środowisku|
|*uv pip install notebook*|Instalacja Jupyter Notebook (jeśli chcesz odpalać lokalnie)|
|*uv sync*|synchronizuje .venv z pyproject.toml|
|*uv add openai*|instaluje i zapisuje w pyproject.toml bibliotekę OpenAI|
